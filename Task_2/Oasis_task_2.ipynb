### Analysing the data before and after lockdown

```python
data_India['day']=data_India['Date'].dt.day
data_India['month']=data_India['Date'].dt.month
data_India['year']=data_India['Date'].dt.year
```

```python
data_b=data_India[(data_India['month']>=1) & (data_India['month']<=4)]
data_b=data_b.groupby('States')['Unemployment rate'].mean().reset_index()
data_b
```

```python
data_a=data_India[(data_India['month']>=4) & (data_India['month']<=7)]
data_a=data_a.groupby('States')['Unemployment rate'].mean().reset_index()
data_a
```

```python
data=data_b
data
```

```python
data['Unemployment rate after lockdown']=data_a['Unemployment rate']
data
```

```python
data['Change in Unemployment rate']=data['Unemployment rate after lockdown']-data['Unemployment rate']/data['Unemployment rate']
```

```python
data
```

```python
fig_diff=px.bar(data_frame=data,x='States',y='Change in Unemployment rate',color='States',title='Change in employment rate')
fig_diff.update_layout(xaxis={'categoryorder':'total descending'})
fig_diff.show()
```

##### Hence we can conclude that the unemployment rates of states Haryana , Jharkhand and Tripura were highest.
